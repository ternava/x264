```python
import matplotlib.pyplot as plt
import numpy as np

all_data = df['Option']
labels1 = '#Location in code'
labels2 = '#Files with an option'

plt.rc('font', size=10) 

fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(6, 3))

ax1.barh(all_data, df['WCount'],align='center')
ax2.barh(all_data, df['FCount'],align='center', color='red')
#ax1.set_yticks(all_data)
#ax2.set_yticks(all_data)
#ax1.set_yticklabels(all_data)
ax2.set_yticklabels('')
#ax1.invert_yaxis()
#ax2.invert_yaxis()

ax1.set_xlabel(labels1)
ax2.set_xlabel(labels2)

fig.tight_layout(pad=0.5)

ax1.grid(which='both', axis='x')
ax2.grid(which='both', axis='x')

plt.savefig("addedloc.pdf", bbox_inches = 'tight', pad_inches = 0)

plt.show()
```
