```python
df = pd.read_csv('/home/xternava/Documents/GitHub/x264/measures/stats_encodingtime2.csv')
df.drop(['Encoding_bitrate'],inplace=True,axis=1)
table = pd.pivot_table(data=df,index=['Preset'], columns=['Specialized_sys'], values=['Encoding_time'], aggfunc=np.mean)
table
#table.plot(kind='bar')
```

```python
df3 = pd.read_csv('/home/xternava/Documents/GitHub/x264/measures/stats_encodingtime3.csv')
df3.drop(['EncodingBitrate'],inplace=True,axis=1)
table = pd.pivot_table(data=df3,index=['Preset'], columns=['System'], values=['EncodingTime'], aggfunc=np.mean)
table
```

```python
table.plot(kind='bar', width=0.7, figsize=(5, 3),  align='center')
plt.ylabel('Encoding time [seconds]')
plt.savefig("encodingtime.pdf", bbox_inches = 'tight', pad_inches = 0)

plt.show()
```

```python
df = pd.read_csv('/home/xternava/Documents/GitHub/x264/measures/stats_encodingtime2.csv')
df.drop(['Encoding_time'],inplace=True,axis=1)
table2 = pd.pivot_table(data=df,index=['Preset'], columns=['Specialized_sys'], values=['Encoding_bitrate'], aggfunc={'Encoding_bitrate':np.mean})
table2
#table.plot(kind='bar')
```
