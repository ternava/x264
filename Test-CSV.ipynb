```python
import matplotlib.pyplot as plt
import numpy as np

import pandas as pd 
df = pd.read_csv('/home/xternava/Documents/GitHub/x264/measures/exesize.csv')
df12= df.drop([0], axis=0)
all_data1 = df12['Binary_size']
labels1 = ['Binary size']
all_data2 = df12['Nr_gadgets']
labels2 = ['Number of gadgets']

fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(10, 3))

# rectangular box plot
bplot1 = ax1.boxplot(all_data1,
                     vert=True,  # vertical box alignment
                     patch_artist=True,  # fill with color
                     labels=labels1)  # will be used to label x-ticks
#ax1.set_title('Binary size')
#ax1.set_xlabel('Binary size')
ax1.scatter(1,3096176, marker ="*", color="blue", s=100)

# notch shape box plot
bplot2 = ax2.boxplot(all_data2,
                     # notch=True,  # notch shape
                     vert=True,  # vertical box alignment
                     patch_artist=True,  # fill with color
                     labels=labels2)  # will be used to label x-ticks
#ax2.set_title('Number of gadgets')
#ax2.set_xlabel('Number of gadgets')
ax2.scatter(1,106495, marker ="*", color="blue", s=100)

# fill with colors
colors = ['pink', 'lightblue', 'lightgreen']
for bplot in (bplot1, bplot2):
    for patch, color in zip(bplot['boxes'], colors):
        patch.set_facecolor(color)

# adding horizontal grid lines
# for ax in [ax1, ax2]:
#    ax.yaxis.grid(True)
#    ax.set_xlabel('Three separate samples')
#    ax.set_ylabel('Observed values')

plt.show()
```
